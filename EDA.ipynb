```python
# data cleaning

data["itemprice"]=data["itemprice"].fillna(55)
data.describe()
```

```python
# data visualization-is whereby you visualize or represent your data in form of graphs

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="gender",data=data, palette="Reds")

# palette colours Reds Greens,dark,bright,purbles,coolwarm,viridis
```

```python
sns.countplot(x="payment",data=data,palette="Greens")
```

```python
sns.countplot(x="product",data=data,palette="Blues")
plt.xticks(rotation=45) 
# to make your x axis to slunt
```

```python
sns.countplot(x="product",data=data,palette="Greens",hue="city")
plt.xticks(rotation=45)
```

```python
# plotting a pie chart

data['payment'].value_counts().plot.pie(autopct='%1.1f%%')
```
