```python
import pandas
data=pandas.read_csv("datasets/shopping_trends.csv")
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Gender",data=data, palette="Reds")
plt.xticks(rotation=90)
```

```python
sns.countplot(x="Category",data=data,palette="Greens",hue="Gender")
plt.xticks(rotation=90)
```
