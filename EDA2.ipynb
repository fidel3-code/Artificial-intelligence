```python
import pandas
data=pandas.read_csv("datasets/sales.csv")
data
```

```python
data.describe()
```

```python
data["quantity"]=data["quantity"].fillna(10)
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="name",data=data, palette="Reds",hue="category")
plt.xticks(rotation=90)
plt.savefig("graph.png")
```

```python
data['category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.savefig("piechart.png")
```
