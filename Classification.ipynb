```python
import pandas
data=pandas.read_csv("datasets/recipe.csv")
data
```

```python
data.describe()
```

```python
data["Sugar"]=data["Sugar"].fillna(23)
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Sugar",data=data, palette="Greens")
plt.xticks(rotation=90)
plt.savefig("sugar.png")
```

```python
data['Vanilla'].value_counts().plot.pie(autopct="%1.1f%%")
plt.savefig("chart.png")
```

```python
# splitting our data into input and output

array=data.values

X=array[0:,0:8]
# X represents the input of the ingredients
# X
# y represents the output or the outcome
Y=array[:,8]
# Y
```

```python
# splitting our data into training data and testing data

from sklearn import model_selection

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, 
                                                                test_size=0.30, 
                                                                   random_state=42)
```

```python
# fit your training data into the model selected

from sklearn.svm import SVC

model=SVC()
model.fit(X_train, Y_train)
```

```python
# testing the model for accuracy
predictions=model.predict(X_test)
print("The prediction is ",predictions)
print("The expected output is ",Y_test)
```

```python
# calculating the accuracy of the model
from sklearn.metrics import accuracy_score
print('Accuracy',accuracy_score(Y_test, predictions))
```

```python
recipe=[[40,27,50,5,10,4,14,6]]
outcome=model.predict(recipe)
print("You will be cooking a ",outcome)
```
